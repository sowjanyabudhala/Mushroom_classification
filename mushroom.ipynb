Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
for column in df.columns:
  df[column]=le.fit_transform(df[column])
```

```python
df.head()
```

store feature matrix in X and response (target)in y

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
X
```

```python
y
```

**Applying PCA**:It is a dimensionality reduction technique that is used to reduce the dimensionality of the large dataset by transforming a large set of variables into a smaller one.By preserving as much as information possible.

```python
from sklearn.decomposition import PCA
pca1=PCA(n_components=7)
pca_fit=pca1.fit_transform(X)
```

```python
pca1.explained_variance_ratio_
```

```python
sum(pca1.explained_variance_ratio_)
```

**conclusion**:Here seven components retaining 85% of the data in te dataset.

Splitting the dataset into training set and test set:we train our models on X_train and y_train.We perform prediction using X_test and we will compare our predictive reuslts by our model with y_test.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(pca_fit,y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

Import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

Model Training

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)

svc=SVC()
svc.fit(X_train,y_train)

dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)

rf=RandomForestClassifier()
rf.fit(X_train,y_train)

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)

```

Prediction on test data

```python
y_pred1=lr.predict(X_test)
y_pred2=knn.predict(X_test)
y_pred3=svc.predict(X_test)
y_pred4=dtc.predict(X_test)
y_pred5=rf.predict(X_test)
y_pred6=gb.predict(X_test)
```

Evaluating the Algorithm

```python
from sklearn.metrics import accuracy_score
```

```python
print('ACC LR',accuracy_score(y_test,y_pred1))
print('ACC KNN',accuracy_score(y_test,y_pred2))
print('ACC SVC',accuracy_score(y_test,y_pred3))
print('ACC DTG',accuracy_score(y_test,y_pred4))
print('ACC RF',accuracy_score(y_test,y_pred5))
print('ACC GB',accuracy_score(y_test,y_pred6))
```

```python
final_data=pd.DataFrame({'models':['LR','KNN','SVC','DTG','RF','GB'],'ACC':[accuracy_score(y_test,y_pred1)*100,
                                                               accuracy_score(y_test,y_pred2)*100,

                                                                 accuracy_score(y_test,y_pred3)*100,
                                                                 accuracy_score(y_test,y_pred4)*100,
                                                                 accuracy_score(y_test,y_pred5)*100,
                                                                 accuracy_score(y_test,y_pred6)*100,]})
```

```python
sns.barplot(x=final_data['models'],y=final_data['ACC'])
```
